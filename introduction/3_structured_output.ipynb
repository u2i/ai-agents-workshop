# Structured Output

## PydanticAI


`PydanticAI` is a powerful library that allows you to get structured, type-hinted output from LLMs, using Pydantic models. This is incredibly useful when you need the model's output to be in a specific format, for example, to populate a database or to feed into another function.

First, let's install it.


```python
OLLAMA_BASE_URL = "http://ollama:11434/v1"
OLLAMA_API = "http://ollama:11434/api/chat"
MODEL = "llama3.2"

headers = {"Content-Type": "application/json"}
```

```python
from pydantic_ai.models.openai import OpenAIChatModel
from pydantic_ai.providers.ollama import OllamaProvider


ollama_model = OpenAIChatModel(
    model_name=MODEL,
    provider=OllamaProvider(base_url=OLLAMA_BASE_URL),
)
```

```python
from pydantic import BaseModel

from pydantic_ai import Agent


class CityLocation(BaseModel):
    city: str
    country: str


agent = Agent(ollama_model, output_type=CityLocation)

result = await agent.run("Where were the olympics held in 2012?")

print(result.output)
```
